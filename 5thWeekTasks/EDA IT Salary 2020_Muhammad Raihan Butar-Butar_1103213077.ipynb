# IMPORT LIBRARY

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# MEMBACA DATASET
Prompt : berikan kode untuk membuat dataframe dari file csv dengan nama file IT Salary Survey EU 2020.csv yang berada di folder Dataset pada google drive pribadi di google colab setelah mount dengan drive


```python
csv_file_path = '/content/drive/My Drive/Dataset/IT Salary Survey EU 2020.csv'

# Load CSV into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
df.head()
```

# MEMBERSIHKAN DATASET
Prompt : beri kode untuk mengubah nilai null dari dataframe saya, jika tipe data float ubah menjadi 0 dan jika tipe data non numeric menjadi "-"


```python
# Fungsi untuk mengubah nilai null
def replace_null(value):
    if pd.isnull(value):
        if isinstance(value, float):
            return 0
        else:
            return "-"
    return value

# Menggunakan applymap untuk menerapkan fungsi ke setiap elemen dalam dataframe
df = df.applymap(replace_null)

df.head()
```

# VISUALISASI DATA

Prompt : berikan kode untuk melakukan visualisasi masing-masing data numerik dengan histogram


```python
for column in df.select_dtypes(include='number').columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column].dropna(), bins=10, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

# DETEKSI OUTLIER
Prompt : berikan kode untuk melakukan deteksi outlier pada kolom Age pada dataframe IT Salary Survey EU 2020.csv lalu tampilkan visualisasinya menggunakan scatter plot dengan membandingkannya dengan kolom Yearly Brutto Salary

```python
# Menghitung IQR pada kolom Age
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

# Menentukan batas bawah dan batas atas untuk outlier
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

# Mendeteksi outlier
outliers_age = df[(df['Age'] < lower_bound_age) | (df['Age'] > upper_bound_age)]

# Visualisasi outlier dengan scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Yearly brutto salary (without bonus and stocks) in EUR'], c='blue', label='Data')
plt.scatter(outliers_age['Age'], outliers_age['Yearly brutto salary (without bonus and stocks) in EUR'], c='red', label='Outliers')
plt.title('Scatter Plot of Age vs Yearly Brutto Salary with Outliers')
plt.xlabel('Yearly brutto salary (without bonus and stocks) in EUR')
plt.ylabel('Age')
plt.legend()
plt.grid(True)
plt.show()
```

# ANALISIS KORELASI
Prompt : berikan kode untuk melakukan analisis korelasi dengan hanya menggunakan data yang bersifat numerik


```python
# Pilih hanya kolom-kolom yang bersifat numerik
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Melakukan analisis korelasi
correlation_matrix = numeric_df.corr()

# Tampilkan matriks korelasi
print("Correlation Matrix:")
print(correlation_matrix)

# Visualisasi matriks korelasi dengan heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

# DISTRIBUSI KELAS
Prompt : bagaimana cara saya melakukan analisis distribusi kelas?

```python
# Analisis distribusi kelas
kelas_distribution = df['Yearly brutto salary (without bonus and stocks) in EUR'].value_counts()

# Tampilkan distribusi kelas
print("Distribusi Kelas:")
print(kelas_distribution)

# Visualisasi distribusi kelas dengan bar plot
plt.figure(figsize=(8, 6))
kelas_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
