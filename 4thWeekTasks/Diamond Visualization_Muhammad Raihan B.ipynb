# 1. BOX PLOT
Prompt : Menggunakan dataframe diatas, berikan kode untuk menampilkan boxplotnya, gunakan ukuran font 11 dan berikan warna biru untuk boxplotnya

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur style
sns.set_style("whitegrid")

# Mengatur ukuran font
sns.set_context("notebook", font_scale=1.1)

# Membuat boxplot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.boxplot(data=df, color='cyan')

# Menambahkan judul
plt.title('Boxplot for Diamond Dataset', fontsize=14)

# Menampilkan boxplot
plt.show()
```

# 2. SCATTER PLOT
Prompt : berikan kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Mengambil subset DataFrame hanya dengan fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Menggunakan pairplot untuk membuat scatter plot untuk setiap pasangan fitur numerik
sns.pairplot(numerical_features)

# Menampilkan plot
plt.show()
```

# 3. VIOLIN PLOT
Prompt : berikan kode untuk menampilkan violin plot untuk semua fitur numerik dengan font 11

```python
sns.set_context("notebook", font_scale=1.1)

# Mengambil subset DataFrame hanya dengan fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Membuat violin plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))  # Menyesuaikan ukuran plot jika diperlukan
sns.violinplot(data=numerical_features, inner="quartile")

# Menambahkan judul
plt.title('Violin Plot for Numerical Features', fontsize=14)

# Menampilkan plot
plt.show()
```

# 4. HEAT MAP
Prompt : berikan kode untuk menampilkan heat map untuk semua fitur numerik

```python
# Mengambil subset DataFrame hanya dengan fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = numerical_features.corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))  # Menyesuaikan ukuran plot jika diperlukan
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")

# Menambahkan judul
plt.title('Heatmap of Correlation Between Numerical Features', fontsize=14)

# Menampilkan plot
plt.show()
```

# 5. LINE PLOT
Prompt : berikan kode untuk menampilkan line plot untuk semua fitur numerik

```python
# Mengambil subset DataFrame hanya dengan fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Mendapatkan daftar nama fitur numerik
features = numerical_features.columns.tolist()

# Membuat line plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))  # Menyesuaikan ukuran plot jika diperlukan
for feature in features:
    plt.plot(df.index, df[feature], label=feature)

# Menambahkan judul dan label
plt.title('Line Plot for Numerical Features', fontsize=14)
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Menampilkan plot
plt.show()
```
