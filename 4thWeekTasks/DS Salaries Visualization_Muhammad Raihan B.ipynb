# 1. BOX PLOT
Prompt : Menggunakan dataframe diatas, berikan kode untuk menampilkan boxplotnya, gunakan ukuran font 11 dan berikan warna cyan untuk boxplotnya

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style seaborn
sns.set(style="whitegrid")

# Membuat boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, color='cyan')

# Mengatur ukuran font
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.xlabel('Features', fontsize=11)
plt.ylabel('Values', fontsize=11)
plt.title('Boxplot of DataFrame', fontsize=13)

# Menampilkan boxplot
plt.show()
```

# 2. SCATTER PLOT
Prompt : berikan kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(df.select_dtypes(include=['int64']), height=2.5)
plt.show()
```

# 3. VIOLIN PLOT
Prompt : berikan kode untuk menampilkan violin plot untuk semua fitur numerik dengan font 11 menggunakan tema berwarna hijau

```python
# Set tema seaborn dan ukuran font
sns.set(style="whitegrid", font_scale=1.1)
sns.set_palette("Greens")

# Membuat violin plot untuk semua fitur numerik
plt.figure(figsize=(10, 6))
sns.violinplot(data=df.select_dtypes(include=['int64']))
plt.title('Violin Plot for Numeric Features', fontsize=13)
plt.xlabel('Features', fontsize=11)
plt.ylabel('Values', fontsize=11)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Menampilkan violin plot
plt.show()
```

# 4. HEATMAP
Prompt : berikan kode untuk menampilkan heat map untuk semua fitur numerik

```python
# Menghitung korelasi antar fitur numerik
correlation_matrix = df.select_dtypes(include=['int64']).corr()

# Membuat heat map
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numeric Features', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

# 5. LINE PLOT
Prompt : berikan kode untuk menampilkan line plot untuk semua fitur numerik

```python
import matplotlib.pyplot as plt

# Mendapatkan list dari fitur numerik
numeric_features = df.select_dtypes(include=['int64']).columns.tolist()

# Membuat line plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
for feature in numeric_features:
    plt.plot(df.index, df[feature], label=feature)

# Menambahkan label dan judul
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot for Numeric Features')
plt.legend()
plt.show()
```
